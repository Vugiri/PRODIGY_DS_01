```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
df = '/content/API_SP.POP.TOTL_DS2_en_csv_v2_543249.csv'
data = pd.read_csv(df, skiprows=4)
```

```python
data
```

```python
population_2020 = data['2020'].dropna()

```

```python
population_2020
```

```python
plt.figure(figsize=(10, 6))
plt.hist(population_2020, bins=30, edgecolor='black')
plt.title('Distribution of Population in 2020')
plt.xlabel('Population')
plt.ylabel('Number of Countries')
plt.grid(True)
plt.show()
```

```python

#bar
import matplotlib.pyplot as plt
population_2020.sort_values(ascending=False).head(10).plot.bar()
plt.title('Top 10 Countries by Population in 2020')
plt.xlabel('Country')
plt.ylabel('Population')
plt.show()
```

```python
#pie
plt.figure(figsize=(8, 8))
plt.pie(population_2020.sort_values(ascending=False).head(10), labels=population_2020.sort_values(ascending=False).head(10).index, autopct='%1.1f%%')
plt.title('Top 10 Countries by Population in 2020')
plt.axis('equal')
plt.show()
```

```python
#scatter plot
plt.scatter(population_2020.index, population_2020)
plt.title('Population in 2020')
plt.xlabel('Country')
plt.ylabel('Population')
plt.show()
```

```python
#boxplot
plt.boxplot(population_2020)
plt.title('Distribution of Population in 2020')
plt.xlabel('Country')
plt.ylabel('Population')
plt.show()
```
